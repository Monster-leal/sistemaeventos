```python
import pandas as pd
from google.colab import files

# Carregar os arquivos do seu computador para o Colab
uploaded = files.upload()

# Listar os arquivos carregados
for filename in uploaded.keys():
    print(f"Arquivo carregado: {filename}")

# Colunas específicas que você quer copiar
colunas_fornecidas = ['Endereço', 'Código ERP', 'Produto', 'Lote', 'Estoque esperado', 'Qtd Validada']

# Carregar os dados das planilhas
planilha_inventario = pd.read_excel('Relatório de Inventário.xlsx')
planilha_teste = pd.read_excel('teste.xlsx')

# Obter os códigos ERP únicos da planilha "teste.xlsx"
codigos_erp_teste = planilha_teste['Código ERP'].unique()

# Filtrar o "Relatório de Inventário.xlsx" para os códigos ERP presentes na planilha "teste.xlsx"
filtro = planilha_inventario['Código ERP'].isin(codigos_erp_teste)
dados_filtrados = planilha_inventario[filtro]

# Verificar se há dados para copiar
if dados_filtrados.empty:
    print("Não há dados para copiar.")
else:
    # Verificar se todas as colunas fornecidas existem no dataframe
    colunas_presentes = [col for col in colunas_fornecidas if col in dados_filtrados.columns]

    if not colunas_presentes:
        print("Nenhuma das colunas fornecidas está presente nos dados filtrados.")
    else:
        # Solicitar o nome da nova planilha
        nome_planilha = input("Digite o nome da planilha para salvar os dados filtrados (sem a extensão .xlsx): ")

        # Selecionar apenas as colunas presentes
        dados_selecionados = dados_filtrados[colunas_presentes]

        # Adicionar os dados selecionados como uma nova planilha em "teste.xlsx"
        with pd.ExcelWriter('teste.xlsx', mode='a', engine='openpyxl') as writer:
            dados_selecionados.to_excel(writer, sheet_name=nome_planilha, index=False)

        print(f"Dados copiados para 'teste.xlsx', na planilha '{nome_planilha}'. Foram adicionadas {dados_selecionados.shape[0]} linhas.")

print("Processo concluído.")

# Baixar o arquivo Excel modificado
files.download('teste.xlsx')
```
